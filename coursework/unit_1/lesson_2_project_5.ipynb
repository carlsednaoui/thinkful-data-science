```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.sql import select
from sqlalchemy import inspect
```

```python
engine = create_engine('sqlite:///airbnb.db')
```

```python
airbnb_data = pd.read_csv('./data/airbnb-listings.csv.gz')
```

```python
airbnb_data.head()
```

```python
airbnb_data.to_sql('listings', engine)
```

```python
inspector = inspect(engine)
for table_name in inspector.get_table_names():
    for column in inspector.get_columns(table_name):
        print("Column: {}, type: {}".format(column['name'], column['type']))
```

```python
pd.read_sql('SELECT * FROM listings LIMIT 3', engine)
```

```python
pd.read_sql('SELECT distinct(cancellation_policy) FROM listings', engine)
```

```python
engine.execute('CREATE index cancellation_policy_index ON listings (cancellation_policy)')
```

```python
pd.read_sql('SELECT distinct(cancellation_policy) FROM listings', engine)
```

```python
pd.read_sql('SELECT cancellation_policy, count(cancellation_policy) FROM listings GROUP BY cancellation_policy ORDER BY 2 DESC', engine)
```
