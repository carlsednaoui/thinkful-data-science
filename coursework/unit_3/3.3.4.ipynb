```python
# Generate a binomial distribution with n=20 and p=0.5.
binomial = np.random.binomial(20, 0.9, 1000)

# Plot a histogram
plt.hist(binomial)

# Print the histogram.
plt.show()
```

```python
# Generate a gamma distribution with shape =5 and scale = 1
gamma = np.random.gamma(78, 1, 100000)

# Plot a histogram.
plt.hist(gamma, bins=100)

# Print the histogram.
plt.show()
```
