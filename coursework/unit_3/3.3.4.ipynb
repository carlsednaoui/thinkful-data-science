```python
# https://courses.thinkful.com/data-201-prepv1/project/3.3.4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate a binomial distribution with n=20 and p=0.5.
binomial = np.random.binomial(20, 0.9, 1000)

# Plot a histogram
plt.hist(binomial)

# Print the histogram.
plt.show()
```

```python
# Generate a gamma distribution with shape =5 and scale = 1
gamma = np.random.gamma(78, 1, 100000)

# Plot a histogram.
plt.hist(gamma, bins=100)

# Print the histogram.
plt.show()
```

```python
# Generate a Poisson distribution with lambda = 3
# lambda is the average number of times the event will occur during the interval
poisson = np.random.poisson(8, 1000)
poisson_mean = np.mean(poisson)
print(poisson_mean)

# Plot a histogram.
plt.hist(poisson)

# Print the histogram.
plt.show()
```

```python
# Assignment

# 1: Create the random var: np.random.distributionname([arguments], 100)
# 2: Graph it
# 3: Get the mean and stddev and create a QQplot
# 4: Where descriptive statistices useful here?

# poisson
poisson = np.random.poisson(10, 1000)
plt.hist(poisson)
plt.show()

poisson_mean = np.mean(poisson)
poisson_std = np.std(poisson)
print(poisson_mean, poisson_std)

# Use the same mean as you did with the poisson distribution
normal = np.random.normal(10, 5, 1000)

poisson.sort()
normal.sort()

plt.plot(normal, poisson, "o")
plt.show()
```

```python
# lognormal
logno = np.random.lognormal(100, 1, 1000)
plt.hist(logno, bins=100)
plt.show()

logno.sort()
plt.plot(normal, logno, "o")
```

```python
# exponential
expo = np.random.exponential(100, 1000)
plt.hist(expo, bins=100)
plt.show()

expo.sort()

plt.plot(normal, expo, "o")
```

```python
# chisquare
chi = np.random.chisquare(10, 1000)
plt.hist(chi)
plt.show()

chi.sort()

plt.plot(normal, chi, "o")
```

```python
# pareto
pare = np.random.pareto(100, 1000)
plt.hist(pare)
plt.show()

pare.sort()

plt.plot(normal, pare, "o")
plt.show
```

```python
# triangular
tri = np.random.triangular(-100, 0, 100, 1000)
plt.hist(tri)
plt.show()

tri.sort()
plt.plot(normal, tri, "o")
plt.show()

plt.plot(normal, normal, "o")
plt.show()
```

```python
# combine graphs
plt.plot(normal, tri, "o")
plt.plot(normal, normal, "o")
plt.legend(['Triangular', 'Normal'], loc='upper left')
plt.show()
```
