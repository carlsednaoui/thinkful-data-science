```python
# Clean text

import string
from nltk.corpus import words
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

# Get stopwords_set and punctuation set.
stopwords = set(stopwords.words('english'))
punctuation = set(string.punctuation)


def clean_text(text):
    word_tokens = word_tokenize(text)
    filtered_sentence = [w.lower() for w in word_tokens if not w in stopwords and not w in punctuation and w.isalpha()]
    return filtered_sentence
```

```python
from collections import Counter

cutoff = 0
all_words = []

for page in all_page_text:
    words = clean_text(page)
    if cutoff > 0:
        all_words.extend(words[:cutoff])
    else:
        all_words.extend(words)

counts = Counter(all_words)
counts.most_common()
```
