```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
urls_to_scrape = pd.read_csv("./data/filtered_domains.csv")
print(urls_to_scrape.shape)
urls_to_scrape.columns
```

```python
homepages = []

for i, r in urls_to_scrape.iterrows():
    print(r.domain_name)
    try:
        page = requests.get("https://www." + r["domain_name"])
        soup = BeautifulSoup(page.content, 'html.parser')
    
        # get the homepage
        homepages.append(soup)
    except:
        homepages.append("NA")

urls_to_scrape["html"] = homepages
```
