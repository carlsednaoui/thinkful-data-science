```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(range(9))
plt.show()
```

```python
np.random.seed(1221) # this makes the random function consistent and replicable
df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)

print(df)
```

```python
plt.plot(df['rand'])
plt.show()
```

```python
plt.plot(df['rand'], color = 'purple')
plt.ylim([-0.1, 1.1])
plt.ylabel("Values")
plt.xlabel("Randomly created number")
plt.title("Random series")
plt.show()
```

```python
plt.plot(df['rand'], color = 'purple')
plt.plot(df['rand_shift'], color = 'green')
plt.ylim([-0.1, 3.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.scatter(x = df['rand'], y = df['rand_sq'])
plt.show()
```

```python
plt.scatter(x = df['rand'], y = df['rand_sq'], color = 'purple', marker = 'x', s = 10)
plt.scatter(x = df['rand'], y = df['rand_shift'], color = 'green', marker = 'x', s = 10)
plt.show()
```

```python
df.plot(kind = 'scatter', x = 'rand',y = 'counts_sqrt')
df.plot(kind = 'line')
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2, 1, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(2, 1, 2)
plt.plot(df['rand_shift'], color='green')
plt.ylabel('Shifted Values')
plt.title('Shifted Series')
plt.show()
```

```python
plt.figure(figsize=(5,10))

plt.subplot(2, 1, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(2, 1, 2)
plt.scatter(x = df['rand_sq'], y = df['rand'], color='green')
plt.ylabel('Squared Values')
plt.title('Squared Series')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(111) #sets the seed for random generation
x = np.random.normal(10, 5, 1000) #Generate 1000 Random Variables
plt.hist(x)
```

```python
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)
plt.hist(x, color='red', bins=np.arange(-10, 40), alpha=.5)  # alpha just controls the opacity
plt.hist(y, color='blue', bins=np.arange(-10, 40), alpha=.5)
plt.title('Histogram Demo')
plt.xlabel('Random Values')
plt.show()
```

```python
# let's normalize this
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)
plt.hist(x, normed=True, color='red', bins=np.arange(-10, 40), alpha=.5) #alpha just controls the opacity
plt.hist(y, normed=True, color='blue', bins=np.arange(-10, 40), alpha=.5)
plt.title('Histogram Demo')
plt.xlabel('Random Values')
plt.show()
```

```python
#you can have as many bins as your heart desires
x = np.random.normal(10, 5, 1000)
plt.hist(x, bins=100, normed=True, color='red')
plt.title('Histogram Demo')
plt.xlabel('Random Values')
plt.show()
```
